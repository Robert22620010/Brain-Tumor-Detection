```python
# Brain Tumor Prediction

# Import Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Load Dataset
df = pd.read_csv("/kaggle/input/brain-tumor/Brain Tumor.csv")
print("Dataset Loaded:")
print(df.head())

# Separate Features and Target
# Drop non-numeric columns
X = df.drop(columns=['Class']).select_dtypes(include=['float64', 'int64'])
y = df['Class']  # Target label

print("\nFeatures used for prediction:")
print(X.columns)

# Handle Missing Values
imp = SimpleImputer(strategy='median')
X_imputed = imp.fit_transform(X)

# Scale Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# Train Random Forest Model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Evaluate Model
y_pred = model.predict(X_test)
print("\nAccuracy on Test Set:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Take User Input (EXCLUDE CLASS)
print("\nEnter values for the following features:")

user_values = []
for col in X.columns:
    val = float(input(f"{col}: "))
    user_values.append(val)

# Convert user input into DataFrame
user_df = pd.DataFrame([user_values], columns=X.columns)

# Impute + Scale
user_array = imp.transform(user_df)
user_array = scaler.transform(user_array)

# Predict Class
prediction = model.predict(user_array)

# Map 0 → No Tumor, 1 → Tumor
class_mapping = {0: 'No Tumor', 1: 'Tumor'}
predicted_label = class_mapping[prediction[0]]

print("\nPredicted Class:", predicted_label)

# Optional: show prediction probability
proba = model.predict_proba(user_array)[0]
print(f"Probability -> No Tumor: {proba[0]:.2f}, Tumor: {proba[1]:.2f}")
```
